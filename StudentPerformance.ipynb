# Analysis

## Correlation Analysis

```python
import matplotlib.pyplot as plt
import seaborn as sns

plt.figure(figsize=(5, 4))

sns.heatmap(df.corr(), annot=True, cmap='coolwarm')

plt.show()
```

```python
plt.figure(figsize=(8,5))
sns.histplot(df['Performance Index'], bins=10, kde=True, color='skyblue')
plt.title('Distribution of Performance Index')
plt.xlabel('Performance Index')
plt.ylabel('Number of Students')
plt.show()
```

```python
plt.figure(figsize=(8,5))
sns.scatterplot(x='Hours Studied', y='Performance Index', data=df, hue='Performance Index', palette='viridis')
plt.title('Hours Studied vs Performance Index')
plt.show()
```

```python
plt.figure(figsize=(8,5))
sns.scatterplot(x='Previous Scores', y='Performance Index', data=df)
plt.title('Previous Scores vs Current Performance Index')
plt.show()
```

```python
plt.figure(figsize=(8,5))
sns.boxplot(x='Sleep Hours', y='Performance Index', data=df)
plt.title('Performance Index by Sleep Hours')
plt.show()
```

```python
plt.figure(figsize=(6,5))
sns.boxplot(x='Extracurricular Activities', y='Performance Index', data=df)
plt.xticks([0,1], ['No', 'Yes'])
plt.title('Impact of Extracurricular Activities on Performance')
plt.show()
```

```python
plt.figure(figsize=(8,5))
sns.scatterplot(x='Sample Question Papers Practiced', y='Performance Index', data=df)
plt.title('Sample Papers Practiced vs Performance')
plt.show()
```

```python
plt.figure(figsize=(8,5))
sns.scatterplot(x='Hours Studied', y='Sample Question Papers Practiced', hue='Performance Index', data=df, palette='coolwarm')
plt.title('Study Hours & Sample Papers Practiced vs Performance')
plt.show()
```

```python
df['Performance Level'] = pd.cut(df['Performance Index'], bins=[0,50,75,100], labels=['Low','Medium','High'])
sns.countplot(x='Performance Level', data=df, palette='Set2')
plt.title('Number of Students by Performance Level')
plt.show()
```

```python
sns.pairplot(df, hue='Extracurricular Activities', palette='Set1')
plt.suptitle('Pairwise Relationships Between Features', y=1.02)
plt.show()
```
